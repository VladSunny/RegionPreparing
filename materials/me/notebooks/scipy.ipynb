```python
import scipy
import numpy as np
```

```python
def f(x):
    return -(45 * x[0] + 62 * x[1] + 38 * x[2])

bounds = scipy.optimize.Bounds(lb = [20, 0, 0], ub = [np.inf, 40, np.inf])
con1 = scipy.optimize.NonlinearConstraint(
    fun = lambda x : x[0] + x[1] + x[2],
    lb = -np.inf,
    ub = 100
)
con2 = scipy.optimize.NonlinearConstraint(
    fun = lambda x: 3 * x[0] + 5 * x[1] + 2 * x[2],
    lb = -np.inf,
    ub = 350
)

result = scipy.optimize.minimize(f, x0 = [0, 0, 0], bounds=bounds, constraints=[con1, con2])
result.fun
```

```python
def f(x):
    return (
        x[0]**2 + 2 * x[1]**2 + x[2]**2 + x[3]**2 +
        3 * x[0] * x[1] * np.cos(x[2]) +
        4 * x[0] * x[2] * np.sin(x[3]) + 
        2 * x[1] * x[3] * np.sin(x[0] + x[2]) +
        5 * np.cos(x[1] + x[3])
    )

bounds = scipy.optimize.Bounds(lb = [0] * 4, ub = [np.pi * 2] * 4)

result = scipy.optimize.differential_evolution(f, x0 = [np.pi] * 4, bounds=bounds)

result.fun.round(6)
```
